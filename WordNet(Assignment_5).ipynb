Import WordNet

```python
from nltk.corpus import wordnet as wn
```

Synonymy

```python
word = 'Happy'

synsets = wn.synsets(word)

synonyms = set()
for syn in synsets:
  for lemma in syn.lemmas():
    synonyms.add(lemma.name())

print('Synonyms: ', synonyms)
```

Antonymy

```python
antonyms = set()

for syn in synsets:
  for lemma in syn.lemmas():
    if lemma.antonyms():
      antonyms.add(lemma.antonyms()[0].name())


print('Antonyms: ', antonyms)
```

Hypernymy

```python
word = 'dog'
synsets = wn.synsets(word)

hypernyms = set()
for syn in synsets:
  for hyper in syn.hypernyms():
    hypernyms.add(hyper.name())

print('Hypernyms: ', hypernyms)
```

Hypernym Tree

```python
def hypernym_tree(synset, level = 0):
  print(' ' * level + synset.name())
  for hyper in synset.hypernyms():
    hypernym_tree(hyper, level + 1)


hypernym_tree(wn.synsets('dog')[0])
```
