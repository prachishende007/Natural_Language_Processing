```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in 2002"
doc = nlp(text)

for ent in doc.ents:
  print(ent.text, ent.label_)
```

Evaluation Metrics

Precision = TP / (TP + FP)

Recall = TP / (TP + FN)

F1 = 2 * (P * R) / (P + R)


```python
TP = 3
FP = 1
FN = 2

precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)

print(precision, recall, f1)
```
